Постройте круговую диаграмму распределения суши на Земле.

```python
import matplotlib.pyplot as plt
%matplotlib inline

Terrain = ['Леса', 'Степи', 'Тундра', 'Пашня']
Popular = [57, 24, 54, 15]

fig, ax = plt.subplots(figsize=(24, 14))
ax.pie(Popular, labels=Terrain, wedgeprops=dict(width=0.5))
```
